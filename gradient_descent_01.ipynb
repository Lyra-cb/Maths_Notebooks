```python
import numpy as np
import matplotlib.pyplot as plt

# Our loss function y = x²
def loss(x):
    return x**2

# Our derivative = 2x
def derivative(x):
    return 2*x

# Gradient descent
def gradient_descent(starting_x, learning_rate, steps):
    x = starting_x
    history = [x]
    
    for i in range(steps):
        grad = derivative(x)
        x = x - (learning_rate * grad)
        history.append(x)
    
    return history

# Run it
history = gradient_descent(starting_x=4, learning_rate=0.1, steps=20)

# Print the journey
for i, x in enumerate(history):
    print(f"Step {i}: x = {x:.4f}, loss = {loss(x):.4f}")
```

```python
# Plot the loss function
x_vals = np.linspace(-5, 5, 100)
y_vals = x_vals**2

fig, ax = plt.subplots(figsize=(10, 6))

# Draw the curve
ax.plot(x_vals, y_vals, 'b-', linewidth=2, label='Loss function (x²)')

# Plot the gradient descent steps
history_loss = [loss(x) for x in history]
ax.scatter(history, history_loss, color='red', zorder=5, label='Gradient descent steps')

# Connect the dots
ax.plot(history, history_loss, 'r--', alpha=0.5)

ax.set_xlabel('x (coefficient value)')
ax.set_ylabel('Loss')
ax.set_title('Gradient Descent rolling down the Loss Function')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Too large learning rate
history_large = gradient_descent(starting_x=4, learning_rate=0.9, steps=20)

for i, x in enumerate(history_large):
    print(f"Step {i}: x = {x:.4f}, loss = {loss(x):.4f}")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Draw the curve
ax.plot(x_vals, y_vals, 'b-', linewidth=2, label='Loss function (x²)')

# Plot both runs
history_loss = [loss(x) for x in history]
history_large_loss = [loss(x) for x in history_large]

ax.scatter(history, history_loss, color='green', zorder=5, label='Learning rate 0.1 (smooth)')
ax.plot(history, history_loss, 'g--', alpha=0.5)

ax.scatter(history_large, history_large_loss, color='red', zorder=5, label='Learning rate 0.9 (bouncing)')
ax.plot(history_large, history_large_loss, 'r--', alpha=0.5)

ax.set_xlabel('x (coefficient value)')
ax.set_ylabel('Loss')
ax.set_title('Gradient Descent — Learning Rate Comparison')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Catastrophic learning rate
history_chaos = gradient_descent(starting_x=4, learning_rate=1.1, steps=10)

for i, x in enumerate(history_chaos):
    print(f"Step {i}: x = {x:.4f}, loss = {loss(x):.4f}")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Draw the curve
ax.plot(x_vals, y_vals, 'b-', linewidth=2, label='Loss function (x²)')

# Smooth descent
history_loss = [loss(x) for x in history]
ax.scatter(history, history_loss, color='green', zorder=5, label='LR 0.1 (smooth)')
ax.plot(history, history_loss, 'g--', alpha=0.5)

# Bouncing
history_large_loss = [loss(x) for x in history_large]
ax.scatter(history_large, history_large_loss, color='orange', zorder=5, label='LR 0.9 (bouncing)')
ax.plot(history_large, history_large_loss, color='orange', linestyle='--', alpha=0.5)

# Chaos — clip to visible range so it doesn't break the plot
history_chaos_clipped = [x for x in history_chaos if abs(x) < 10]
history_chaos_loss_clipped = [loss(x) for x in history_chaos_clipped]
ax.scatter(history_chaos_clipped, history_chaos_loss_clipped, color='red', zorder=5, label='LR 1.1 (exploding!)')
ax.plot(history_chaos_clipped, history_chaos_loss_clipped, 'r--', alpha=0.5)

ax.set_ylim(0, 30)
ax.set_xlim(-6, 6)
ax.set_xlabel('x (coefficient value)')
ax.set_ylabel('Loss')
ax.set_title('Gradient Descent — Learning Rate Comparison')
ax.legend()

plt.tight_layout()
plt.show()
```
