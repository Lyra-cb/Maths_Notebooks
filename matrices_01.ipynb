```python
import numpy as np
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
```

```python
print("Shape of A:", A.shape)
print("Shape of B:", B.shape)
```

```python
result = np.dot(A, B)
print("A × B =")
print(result)
```

```python
# A non-square example
C = np.array([[1, 2, 3],
              [4, 5, 6]])   # 2x3

D = np.array([[7,  8],
              [9,  10],
              [11, 12]])    # 3x2

result2 = np.dot(C, D)
print("Shape of C:", C.shape)
print("Shape of D:", D.shape)
print("Result shape:", result2.shape)
print(result2)
```

```python
# 3 patients, each with 3 risk factors (age_scaled, sbp_scaled, smoking)
patients = np.array([[0.8, 1.2, 1.0],
                     [0.5, 0.9, 0.0],
                     [1.1, 1.4, 1.0]])

# Coefficients for each risk factor
coefficients = np.array([0.5, 0.3, 0.8])

# One matrix multiplication → risk scores for ALL patients simultaneously
risk_scores = np.dot(patients, coefficients)
print("Risk scores:", risk_scores)
```

```python
import numpy as np

B = np.array([[2, 1],
              [5, 3]])

B_inv = np.linalg.inv(B)

print("B inverse:")
print(B_inv)

print("\nB × B_inv (should be identity):")
print(np.round(np.dot(B, B_inv)))
```

```python
import numpy as np

D = np.array([[3, 6],
              [2, 4]])

print("Determinant:", np.linalg.det(D))
print("Is singular?", np.isclose(np.linalg.det(D), 0))

try:
    D_inv = np.linalg.inv(D)
    print(D_inv)
except np.linalg.LinAlgError as e:
    print("Error:", e)
```
